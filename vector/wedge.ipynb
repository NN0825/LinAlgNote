```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
```

```python
def create_wedge(ax, radius, height, theta1, theta2, color='blue'):
    # Create the top and bottom faces
    theta = np.linspace(np.radians(theta1), np.radians(theta2), 100)
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    z_top = height * np.ones_like(x)
    z_bottom = np.zeros_like(x)

    # Create the sides
    verts = []
    verts.append(list(zip(x, y, z_top)))
    verts.append(list(zip(x, y, z_bottom)))
    verts.append([(0, 0, height), (x[0], y[0], height), (x[-1], y[-1], height)])
    verts.append([(0, 0, 0), (x[0], y[0], 0), (x[-1], y[-1], 0)])
    verts.append([(0, 0, 0), (0, 0, height), (x[0], y[0], height), (x[0], y[0], 0)])
    verts.append([(0, 0, 0), (0, 0, height), (x[-1], y[-1], height), (x[-1], y[-1], 0)])

    # Create the 3D polygon and add it to the axes
    poly = Poly3DCollection(verts, facecolors=color, linewidths=1, edgecolors='black', alpha=0.5)
    ax.add_collection3d(poly)

# Create a figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the parameters for the wedge
radius = 1.0
height = 1.0
theta1 = 0
theta2 = 90

# Create the wedge
create_wedge(ax, radius, height, theta1, theta2)

# Set the aspect of the plot to be equal
ax.set_box_aspect([1, 1, 1])

# Set the x, y, and z limits of the plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(0, 1.5)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Wedge Plot')

# Display the plot
plt.show()
```
